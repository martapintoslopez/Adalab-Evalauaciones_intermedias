- Haciendo el anÃ¡lisis exploratorio te deberÃ­as haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberÃ­an. Esto es debido a que los decimales estÃ¡n establecidos como comas y no con puntos. Crea una funciÃ³n que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.
    - DespuÃ©s de haber hecho los cambios, chequea los tipos de datos. Â¿Son ya del tipo correcto? En caso de que no, crea otra funciÃ³n o modifica la anterior para que sean de tipo float.
- Hay algunas canciones cuyo tÃ­tulo estÃ¡ en mayÃºsulas. Crea una funciÃ³n para que todos los tÃ­tulos estÃ©n en minÃºscula. Haz lo mismo para las columnas de "title" y "album".
- Algunos de los artistas tienen sÃ­mbolos raros, en concreto el sÃ­mbolo $. Crea una funciÃ³n que los reemplaze pos "s".

```python
#Los nombres de las columnas empiezan con mayÃºsculas, pon todos los nombres de las columnas en minÃºsculas.
new_columns={columna: columna.lower() for columna in df_spotify.columns} 
new_columns
```

```python
#Aplicamos nuevos nombres
df_spotify.rename(columns = new_columns, inplace = True)
df_spotify.head(2)
```

```python
#Crea una *funciÃ³n* que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.
def transform_datos(dato):
    try:
        #Hacemos cambio de comas a puntos y convertimos a float
        return float(dato.replace(",","."))
    except:
        # Si los datos ya han sido cambiados nos los devuelve igual.
        return dato

#Creamos una lista para iterar por las columnas que deben ser modificadas
list_columns_to_change = ["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]
for column in list_columns_to_change:
    #El mÃ©todo .apply pasa por cada uno de los elementos de la columna sobreescribiendo el nuevo valor
    df_spotify[column] = df_spotify[column].apply(transform_datos)
```

```python
#Â¿CuÃ¡les son los valores Ãºnicos de la columna "album_type"? Crea tres dataframes diferentes, uno para cada tipo de "album_type". 
#Â¿CuÃ¡ntas canciones tenemos en cada tipo?
df_spotify["album_type"].unique()
```

```python
# Crear DataFrames separados para cada tipo de "album_type"
df_album = df_spotify[df_spotify["album_type"] == 'album']
df_album.sample(2)
```

```python
df_single = df_spotify[df_spotify["album_type"] == 'single']
df_single.sample(2)
```

```python
df_compilation = df_spotify[df_spotify["album_type"] == 'compilation']
df_compilation.head(2)
```

```python
# El nÃºmero de canciones en cada tipo
print(f"Canciones en los Ã¡lbumes: {len(df_album)}")
print(f"Canciones en los singles: {len(df_single)}")
print(f"Canciones las compilaciones: {len(df_compilation)}")
```

```python
# Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. Â¿CuÃ¡l es la media y la desviaciÃ³n estÃ¡ndar de "danceability",
# "acousticness" y "speechiness" de cada artista?. Muesrtra los resultados en tres dataframes diferentes, uno para cada mÃ©trica ("danceability",
# "acousticness" y "speechiness"). Ordena los resultados de mayor a menor en base a la media.
danceability = df_album.groupby("artist")["danceability"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)
acousticness = df_album.groupby("artist")["acousticness"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)
speechiness = df_album.groupby("artist")["speechiness"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)

print("danceability".upper(),"\n",danceability)
print("---------------")
print("acousticness".upper(),"\n",acousticness)
print("---------------")
print("speechiness".upper(),"\n",speechiness)
print("---------------")
```
