```python
#Si queremos reorganizar las columnas usamos .reindex(columns = [""])
# y en la lista metemos el nuevo orden de las columnas
cols = ['artist', 'url_spotify', 'track', 'album', 'album_type', 'uri',
       'danceability', 'energy', 'key', 'loudness', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'url_youtube', 'title', 'channel', 'views', 'likes',
       'comments', 'description', 'licensed', 'official_video', 'stream']
cols.sort() #Ordenadas alfabéticamente
df_sort = df_spotify.reindex(columns= cols)
df_sort.head(3)
```

```python
#Crea una *función* que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.
def transform_datos(dato):
    try:
        #Hacemos cambio de comas a puntos y convertimos a float
        return float(dato.replace(",","."))
    except:
        # Para que me ponga nulos cuando un dato no sea numérico.
        return np.nan

#Creamos una lista para iterar por las columnas que deben ser modificadas
list_columns_to_change = ["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]
for column in list_columns_to_change:
    #El método .apply pasa por cada uno de los elementos de la columna sobreescribiendo el nuevo valor
    df_spotify[column] = df_spotify[column].apply(transform_datos)
```

```python
#¿Cuáles son los valores únicos de la columna "album_type"? Crea tres dataframes diferentes, uno para cada tipo de "album_type". 
#¿Cuántas canciones tenemos en cada tipo?
df_spotify["album_type"].unique()
```

```python
# Crear DataFrames separados para cada tipo de "album_type"
df_album = df_spotify[df_spotify["album_type"] == 'album']
df_album.sample(2)
```

```python
df_single = df_spotify[df_spotify["album_type"] == 'single']
df_single.sample(2)
```

```python
df_compilation = df_spotify[df_spotify["album_type"] == 'compilation']
df_compilation.head(2)
```

```python
# El número de canciones en cada tipo
print(f"Canciones en los álbumes: {df_album.shape[0]}") # mejor que el que he usado de len(df_album)
print(f"Canciones en los singles: {df_single.shape[0]}")
print(f"Canciones las compilaciones: {df_compilation.shape[0]}")
```

```python
# Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¿Cuál es la media y la desviación estándar de "danceability",
# "acousticness" y "speechiness" de cada artista?. Muesrtra los resultados en tres dataframes diferentes, uno para cada métrica ("danceability",
# "acousticness" y "speechiness"). Ordena los resultados de mayor a menor en base a la media.
danceability = df_album.groupby("artist")["danceability"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)
acousticness = df_album.groupby("artist")["acousticness"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)
speechiness = df_album.groupby("artist")["speechiness"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)

print("danceability".upper())
display(danceability)
print("---------------")
print("acousticness".upper())
display(acousticness)
print("---------------")
print("speechiness".upper())
display(speechiness)
print("---------------")
#Podríamos haber hecho esto con un for y una lista
```

```python
#Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de 
#streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: ¿Son aquellos artistas
#que tienen más visitas los que más likes tienen? Para solucionar este ejercicio deberéis:
#Agrupa por artista y calcular la media de visitas (views) y de likes.
views_mean = df_album.groupby("artist")["views"].agg("mean").sort_values(ascending=False)
likes_mean = df_album.groupby("artist")["likes"].agg("mean").sort_values(ascending=False)
#Quedate con los 10 artistas que más visitas han tenido.
top_10_views = views_mean.head(10)
print(f"Los 10 artistas con las medias de visitas más altas son: {top_10_views}")
#Quedate con los 10 artistas que más likes han tenido.
top_10_likes = likes_mean.head(10)
print(f"Los 10 artistas con las medias de likes más altas son: {top_10_views}")
```
